## Referencing to Outside (above) the Root

Okay So in the root directory of the source/build folders for the sphinx project you have an index file that is supposed to be root, i.e. everything else should be in the same directory or directories below it. At least this what it apeared to me then.

Now my question was "Can sphinx link to documents that are not located in directories below the root document?"
or "how to give a reference to a html file in sphinx"


Okay I found this thread very relevant:

https://stackoverflow.com/questions/10199233/can-sphinx-link-to-documents-that-are-not-located-in-directories-below-the-root


Then trying it out, I realized that it is good for the rst files outside the root, and does not work for htmls ones. 


Also I think these ones are related ( but I did not really read through it to see how it works)

https://stackoverflow.com/questions/17086273/specifying-a-relative-path-in-sphinx-conf-py-for-collaborative-documentation/17121950

https://github.com/sphinx-doc/sphinx/issues/701



Then I thought maybe I should treat it as an external link. I saw before that you can include a general link in your rst.
The syntax was [here](https://sublime-and-sphinx-guide.readthedocs.io/en/latest/references.html)

So I tried this:
```

`Back to Home <../../../index.html>`_

```

in the index.rst of the subpage, and it worked. The key note here is that we treat it like a normal webpage, but the relative address would be calculated based on the location of the build files in the sphinx directory. So in my case it is like 'root/sphinx/build/html/index.html, so that leads to 3 levels of going up in the relative path.

# Customizing the sphinx theme

So the next item on my list was "How I can make this theme a bit different?"  

What I wanted to do here actually was to make the sphinx (read the doc theme) page to have things like: a logo (with link) to my Github page in the sidebar, link to my homepage in the sidebar, and so on.

So turns out that you can modify the read-the-doc theme 
by setting the theme "configuration" in the 'conf.py' file.

Here is the related link 

https://sphinx-rtd-theme.readthedocs.io/en/stable/configuring.html#confval-logo_only

Also on the topic of editing the configuration I found this link also relevant:

https://www.sphinx-doc.org/en/master/usage/configuration.html#confval-html_logo


There is a 'html_theme_options' in the conf.py which you can set some options. But the things you can do here (at least for the rth theme) is not much. 

So I found that among the different options, one thing I could do is that to replace the home icon at the top of the sidebar with a custom icon. I used a small photo of myself as the new icon!

to do that you would set the address of the icon image in the conf.py file like this:
```

html_logo = "../../images/RIMG4.png"
```

then to remove the description from below the logo would set 
```

'logo_only': True

```

this option in the "html_theme_options".


For the "rth" theme, I could not find an easy way to do the other things like modyfiying the sidebar.

I found this thred relevant for more customization of the theme:

https://stackoverflow.com/questions/44793811/change-the-colors-of-the-sphinx-read-the-docs-theme

https://docs.readthedocs.io/en/latest/guides/adding-custom-css.html

https://github.com/readthedocs/sphinx_rtd_theme/issues/777

But at the end, what I realized is that the alabaster theme which I did not like much in the begining is somewhat more flexible in editing the theme.

## Customizing the 'alabastar' Theme

okay giving up on the Read the Docs theme, I went on looking at what you can do for other themes like [basicstrap](https://pythonhosted.org/sphinxjp.themes.basicstrap/design.html)
and [alabaster](https://alabaster.readthedocs.io/en/latest/installation.html).

for the alabaster theme I found these threads very relevant:

https://alabaster.readthedocs.io/en/latest/customization.html

https://stackoverflow.com/questions/33029008/how-to-edit-sidebar-under-sphinx-alabaster-theme

https://alabaster.readthedocs.io/en/latest/installation.html

Looks like for editing the sidebar alabaster gives more flexibility (for what I saw)

We can customize what we put or remove from there:

```
html_sidebars = {
    '**': [
        'about.html',
        'localtoc.html',
        'globaltoc.html',
        'navigation.html',
        'relations.html',
        'searchbox.html',
        'donate.html',
    ]
}
```

so for example I removed both 'searchbox' and 'about' from the sidebar as well as 'blobaltoc'.

Also we can modify the theme options in the 'conf.py' like:

```

html_theme_options = {
    'logo': 'logo.png',
    'github_user': 'bitprophet',
    'github_repo': 'alabaster',
}

```

Default values can be found by viewing theme.conf directly.
https://github.com/bitprophet/alabaster/blob/master/alabaster/theme.conf

I could still set my custom logo with:
```

html_logo = "../../images/RIMG4.png"

```
(I did not make much use of the logo option in  the config, but it seems as a good alternative)


The good part about this html sidebars is that you can include your own sidebar items as described [here](https://stackoverflow.com/questions/33029008/how-to-edit-sidebar-under-sphinx-alabaster-theme)

```

html_sidebars = {
        '**': [
                 'localtoc.html',
                 'relations.html',
                 'searchbox.html',
                 # located at _templates/
                 'foo.html',
            ]

        }
        
```

and we put the special item in the sphinx _template directory. For my case for example I added this one:  'home.html'

```
<hr />
<p>
<h2 class='logo'>
<a href='../../../index.html'>  <img  alt="" height="27" src="../../../images/homeg.png" width="24"> Home </a>
</h2>
</p>

```
