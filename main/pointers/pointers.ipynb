```
class MciWrap(object):
  def __init__(self, data, k):
    self._data = data
    self._k = k

    print(self._data)
    print(self._k)
  @property
  def value(self):
    return self._data[self._k]
  @value.setter
  def value(self, v):
    self._data[self._k] = v

class Form(object):
  def __init__(self, fields):
    self.fields=fields

  @property
  def data(self):
      return self.__dict__["fields"]

  def __getattr__(self, name):
       return MciWrap(self.data, name)

a=Form( [{"id":"fname", 'value':"George"}, {"id":"lname", 'value':'Lucas"'} ])

# a.data

# getattr(a,fields)
a.data
a.fname.value
```
